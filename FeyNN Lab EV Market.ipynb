```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df2['Profession']=lb.fit_transform(df2['Profession'])
df2['Marrital Status']=lb.fit_transform(df2['Marrital Status'])
df2['Education']=lb.fit_transform(df2['Education'])
df2['Personal loan']=lb.fit_transform(df2['Personal loan'])
```

```python
df2.head()
```

```python
from sklearn.cluster import KMeans
```

```python
X=df1[['PriceEuro','Range_Km']].values
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

```python
y_kmeans = kmeans.labels_
y_kmeans
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('PriceEoru')
plt.ylabel('Range_Km')
plt.legend()
plt.show()
```

```python
df1['ClusterLabel'] = y_kmeans+1
```

```python
df1.head()
```

```python
drop=['PowerTrain','PlugType','BodyStyle','Segment','Model','FastCharge_KmH']
df1=df1.drop(drop,axis=1)
```

```python
df1['Brand']=lb.fit_transform(df1['Brand'])
```

```python
df1
```

```python
df1['RapidCharge'].replace('Yes',1,inplace = True)
df1['RapidCharge'].replace('No',0,inplace = True)
```

```python
df1.info()
```

```python
#Using PCA alongs the columns
from sklearn.decomposition import PCA
from sklearn import preprocessing

pca = preprocessing.scale(df1)

pca = PCA(n_components=2)
pc = pca.fit_transform(df1)
```

```python
name=['pc1','pc2']
pcc=pd.DataFrame(data=pc,columns=name)
pcc
```

```python
df1["cluster_num"]=kmeans.labels_
sns.scatterplot(data=pcc, x="pc1", y="pc2", hue=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], 
            marker="X", c="r", s=80, label="centroids")
plt.grid(True)
plt.legend()
plt.show()
```
